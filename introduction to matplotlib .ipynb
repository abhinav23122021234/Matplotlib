## matplotlib example workflow

```python
# 0 . import matplotlib and get it ready for plotting in jupyter 
%matplotlib inline
import matplotlib.pyplot as plt

# 1 . prepare data
x= [1,2,3,4]
y = [11,22,33,44]
# 2 . setup plot 
fig, ax=plt.subplots(figsize=(10,10)) #(width,height)
# 3 . plot data
ax.plot(x,y)

# 4 . customixe plot
ax.set(title="simple plot",
xlabel="x-axis",
ylabel="y-axis")
# 5 save and shhow (you save the whole figure )
fig.savefig("images/sample-plot.png")

```

## two options for subplot

```python
# subplot option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
# plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(100),np.random.random(100))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
#subplots option 2
fig,ax=plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))
# print on basis of indexs starting from 0,0 (row,column)
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(110),np.random.random(110));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(100));
#using semicolon helps in ignoring all unneccesary info in output box
```

### which one should we use ? 
  *when plotting something quickly,okay to use the pyplot method
  
  *when plotting something advance ,use the OO method

```python
over_50=heart_disease[heart_disease["age"] > 50 ]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target')
```

```python
# OO method fixed with pyplot
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)
##ax.set_xlim([45,100]), x ko 45 to 100 show krta h
```

```python
plt.style.use('classic')
## OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

# plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

#customize the plot
ax.set(title="heart diseas and choolestrol",
       xlabel="age",
       ylabel="cholestrol")

#add a legend 
ax.legend(*scatter.legend_elements(),title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
over_50.head()
```

# subplot of chol,age,thalach(max heart rate)

```python
fig,(ax0, ax1 )=plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                            sharex=True)


#add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
#customize ax0
ax0.set(title="heart disease and cholestrol",
        ylabel="cholestrol"
       )
# add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="target");

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")


#add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])


#customizing ax1

ax1.set(title="max heart rate and age",
        xlabel="age",
        ylabel="max heart rate")

#add a legend
ax1.legend(*scatter.legend_elements(),title="target");

#add meanline

ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");
# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight="bold");
```

```python
# customizing the y and x axis limitations
plt.style.use('ggplot')



fig,(ax0, ax1 )=plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                            sharex=True)


#add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")#cmap used to change color of legend
#customize ax0
ax0.set(title="heart disease and cholestrol",
        ylabel="cholestrol"
       )

#change x limit 
ax0.set_xlim([50,80])

#change y limit
ax0.set_ylim([100,500])

# add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="target");

#add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")


#add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                   cmap="winter")


#customizing ax1

ax1.set(title="max heart rate and age",
        xlabel="age",
        ylabel="max heart rate")

#set x limit
#already shared to above value as we used sharex(True)

#set y limit 
ax1.set_ylim([80,200])

#add a legend
ax1.legend(*scatter.legend_elements(),title="target");

#add meanline

ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");
# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight="bold");
plt.savefig('heart disease analysis.pdf')
```
